Questão 2
Enunciado: Crie uma lista com 5 números inteiros. Use um loop for para imprimir o dobro de cada número.

```python
# Resolução
numeros = [2, 5, 10, 21, 50]

print("O dobro de cada número da lista é:")
for n in numeros:
    dobro = n * 2
    print(dobro)
```

Questão 3
Enunciado: Crie um dicionário que represente um carro (marca, modelo, ano). Verifique se o ano é anterior a 2010 e imprima "Carro antigo" ou "Carro novo".

```python
# Resolução
carro = {
    "marca": "Volkswagen",
    "modelo": "Gol",
    "ano": 2008
}

# Verificação lógica usando a chave 'ano'
if carro["ano"] < 2010:
    print(f"O {carro['modelo']} é um carro antigo.")
else:
    print(f"O {carro['modelo']} é um carro novo.")

#Mude os valores das chaves de carro para testar as condições!
```

Questão 4
Enunciado: Baseado no arquivo dados_produtos.csv, crie uma lista com os preços: [85.50, 220.00, 890.00, 150.00, 180.00]. Escreva um programa que calcule a soma total desses preços e a média de valor dos produtos.

```python
precos = [85.50, 220.00, 890.00, 150.00, 180.00]

soma_total = sum(precos)
media_precos = soma_total / len(precos)

print(f"O valor total do estoque é: R$ {soma_total:.2f}")
print(f"A média de preço por produto é: R$ {media_precos:.2f}")
```

Questão 5
Enunciado: Crie uma estrutura condicional para um sistema de desconto. Se o produto for da categoria "Monitores", aplique 10% de desconto no preço (890.00) e exiba o novo valor. Caso contrário, exiba "Sem desconto para este item".

```python
produto_nome = "Monitor 24 pol"
produto_preco = 890.00
categoria = "Monitores"

if categoria == "Monitores":
    desconto = produto_preco * 0.10
    preco_final = produto_preco - desconto
    print(f"O produto {produto_nome} recebeu desconto! Preço final: R$ {preco_final:.2f}")
else:
    print("Sem desconto para este item.")
```
