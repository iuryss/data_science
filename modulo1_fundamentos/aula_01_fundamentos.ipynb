Módulo 1: Fundamentos de Python para Ciência de Dados
Nesta aula, vamos aprender a base da linguagem Python. O objetivo é entender como o computador armazena dados e como podemos dar ordens lógicas a ele.

```python
# 1. Variáveis e Tipos de Dados
nome = "Estudante"          # String (texto)
idade = 20                  # Integer (inteiro)
altura = 1.75               # Float (decimal)
estuda_python = True        # Boolean (verdadeiro/falso)

print(f"Olá, meu nome é {nome} e tenho {idade} anos.")
```

2. Estruturas de Dados
Listas são conjuntos ordenados de itens. Dicionários guardam pares de "Chave" e "Valor".

```python
# Criando uma lista de frutas
frutas = ["Maçã", "Banana", "Uva"]
print(f"Primeira fruta: {frutas[0]}")

# Criando um dicionário de um aluno
aluno = {
    "nome": "João",
    "nota": 8.5
}
print(f"O aluno {aluno['nome']} tirou nota {aluno['nota']}")
```

3. Condicionais e Loops
O if decide se algo acontece. O for repete uma ação.

```python
# Verificando a nota
nota = 7.5
if nota >= 7:
    print("Aprovado!")
else:
    print("Recuperação.")

# Percorrendo a lista de frutas
print("Minha lista de compras:")
for fruta in frutas:
    print(f"- {fruta}")
```
